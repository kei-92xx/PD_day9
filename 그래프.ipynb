## 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
# 지불금액, 팁, 성별, 흡연여부, 요일, 시간, 전체인원
tips= sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
#fig(기본틀), 그래프격자(axes1)
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

## 1. 히스토그램은 df의 열데이터 분포와 빈도를 살펴보는 용도로 자주 사용
### 변수를 하나만 사용해서 그린 '일변량 그래프'

```python
axes1.hist(tips['total_bill'], bins=10) #bins=10 : x축의 간격을 10으로 좆정
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
fig
```

## 2.산점도 그래프
### 변수 2개 사용한 '이변량 그래프'-total_bill열에 따른 tip 열의 분포를 나타낸 산점도 그래프

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 3.박스 그래프
* 이산형 변수-Female, Male처럼 명확하게 구분되는 값을 의미
* 연속형 변수-Tip와 같이 명확하게 셀 수 없는 변위의 값을 의미

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
     tips[tips['sex'] == 'Male']['tip']],
    labels=['Female', 'Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프
* 다변량 - 3개 이상의 변수를 사용

```python
#성별을 새변수로 추가
#(문자열은 산점도 그래프의 색상을 지정하는 값을 사용할 수 없다.)
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips
```

```python
scatter_plot =plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] * 10,
    c=tips['sex_color'],
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 단변량그래프그리기 - 히스토그램

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

## seaborn 라이브러리로 히스토그램
* seaborn 라이브러리로 히스토그램을 그리려면 subplots, distplot 메서드를 사용
* distplot메서드:기본 틀 만듬/total_bill메서드:total_bill열데이터를 전달

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram sith Density Plot')
```

```python
#밀집도(정규화시켜 넓이가 1이 되도록 그린 그래프)제외
#kernel Density Estimation(커널밀도추정)
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# rug인자 추가 - 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집정도를 표현한 그래프
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

### count 그래프-이산값을 나타낸 그래프
* 주로 범주형 변수의 분포를 파악할 때 사용
* sns.countplot(x or y, data)

```python
ax = plt.subplots()
ax = sns.countplot(x='day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.countplot(y='day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

# 다양한 종류의 이변량 그래프 그리기

## 1. seaborn 라이브러리로 산점도 그래프 그리기
### *.regplot메서드를 사용하면 산점도 그래프와 회귀선을 그릴 수 있다.

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

## 바 그래프 그리기

```python
# barplot는 지정한 변수의 평균을 계산하여 그린다.
#시간에 따라 지불한 비용의 평균을 바 그래프로 나타냄
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

## 박스그래프 그리기

```python
# 박스 그래프는 최솟값, 1사분위수, 중간값, 3사분위수, 최대값, 이상치 등
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex',
                    data=tips, split=True)
plt.show()
```
