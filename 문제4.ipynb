### 연령별 통계

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from matplotlib import font_manager, rc
f_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
df = pd.read_excel("../data/seoul.xls")
df.head()
```

```python
data = df.loc[df['대분류'] == '연령별']
data.index = ['10대', '20대', '30대', '40대', '50대', '60대 이상']
data
```

```python
f, ax = plt.subplots(1,3, figsize=(16,8))
explode=[0,0.02,0.02,0.02,0.02,0.02]

data['TV  또는 비디오시청'].plot.pie(ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('연령별 TV TV 또는 비디오시청 통계')
ax[0].set_ylabel('')

data['종교활동'].plot.pie(ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('연령별 종교활동 통계')
ax[1].set_ylabel('')

data['문화예술관람'].plot.pie(ax=ax[2], autopct='%1.1f%%')
ax[2].set_title('연령별 문화예술관람 통계')
ax[2].set_ylabel('')

plt.show()
```

```python
data2 = data.drop(['기간', '대분류', '분류', '합계'], axis=1)
data2.index = ['10대', '20대', '30대', '40대', '50대', '60대 이상']
data3 = data2.T
data3
```

```python
f, ax = plt.subplots(2,3, figsize=(16,8))
ax=ax.ravel()
explode=[0,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02]

data3['10대'].plot.pie(ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('10대의 여가활동 통계')
ax[0].set_ylabel('')

data3['20대'].plot.pie(ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('20대의 여가활동 통계')
ax[1].set_ylabel('')

data3['30대'].plot.pie(ax=ax[2], autopct='%1.1f%%')
ax[2].set_title('30대의 여가활동 통계')
ax[2].set_ylabel('')

data3['40대'].plot.pie(ax=ax[3], autopct='%1.1f%%')
ax[3].set_title('40대의 여가활동 통계')
ax[3].set_ylabel('')


data3['50대'].plot.pie(ax=ax[4], autopct='%1.1f%%')
ax[4].set_title('50대의 여가활동 통계')
ax[4].set_ylabel('')


data3['60대 이상'].plot.pie(ax=ax[5], autopct='%1.1f%%')
ax[5].set_title('60대의 여가활동 통계')
ax[5].set_ylabel('')


plt.show()
```
